```python
import pandas as pd

# 1. Pehle internet se data uthao (Thoda time lagega)
url = 'https://raw.githubusercontent.com/ZhaoyiW/YouTube-Trending-Video-Analysis/master/USvideos.csv'
print("Downloading data... please wait.")
youtube_df = pd.read_csv(url)

# 2. Ab isko apne computer par save karlo
# Yeh file wahan save hogi jahan tumhara Jupyter Notebook chal raha hai
youtube_df.to_csv('USvideos_Local.csv', index=False)

print("Success! File download hokar save ho gayi hai.")
```

```python
youtube_df = pd.read_csv('USvideos_Local.csv')
youtube_df.head()
```

```python
youtube_df.info()
```

```python
youtube_df['trending_date'] = pd.to_datetime(youtube_df['trending_date'],format = '%y.%d.%m')
youtube_df.info()
```

```python
top_views = youtube_df.sort_values(by = 'views',ascending = False)
top_views.head()
```

```python
sorted_column = top_views[['title','views']]
sorted_column.head()
```

```python
top_channel = youtube_df.groupby('channel_title')['views'].sum()

top_news = top_channel.sort_values(ascending = False)
top_news.head()
```

```python
import matplotlib.pyplot as plt

top_5 = top_news.head(5)

top_5.plot(kind = 'bar', color = 'skyblue' , figsize = (10,5))
plt.title('Top 5 youtube channel')
plt.xlabel('Channel name')
plt.ylabel('Total Views')
plt.show()
```

```python
youtube_df.plot(kind = 'scatter',x='views',y='likes',color='red')
plt.show()
```

```python
youtube_df['title_length'] = youtube_df['title'].str.len()
youtube_df.head()
```

```python
youtube_df['title_length'].plot(kind='hist', bins=50 , color='orange', edgecolor= 'white', figsize = (10,5))
plt.title('Distribution title length')
plt.xlabel('Character count')
plt.ylabel('Numbers of videos')
plt.show()                                
                                
                                
```

```python
numeric_cols = ['views','likes','dislikes','comment_count']
correlation_table = youtube_df[numeric_cols].corr()
print(correlation_table)
```

```python
# Clean data ko wapas CSV file mein save karo
# index=False ka matlab hai humein 0,1,2,3 wali numbering nahi chahiye
youtube_df.to_csv('Final_YouTube_Report.csv', index=False)

print("Mubarak ho! File save ho gayi hai.")
```
